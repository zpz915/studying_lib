```python
optimizer = tf.keras.optimizers.Adam()
```

```python
def train_step(model,images,labels):
    with tf.GradientTape() as t:  #记录梯度
        pred =model(images)
        loss_step = tf.keras.losses.BinaryCrossentropy(from_logits=True)(labels,pred)  #因为没有激活所以设为true
    grads = t.gradient(loss_step,model.trainable_variables) #计算梯度 损失以及可训练的参数
    optimizer.apply_gradients(zip(grads,model.trainable_variables))  #优化 
```
