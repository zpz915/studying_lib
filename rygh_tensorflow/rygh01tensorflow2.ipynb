```python
x = data.Education
y = data.Income
```

```python
x
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",
              loss="mse"
)
```

```python
history = model.fit(x,y,epochs=4000)
```

```python
model.predict(x)
```
